```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(1,3,figsize=((15,5)))

ax[0].plot(np.arange(0,5),[0]*5, "o")
ax[0].set_xlabel("ULA")
ax[0].get_yaxis().set_visible(False)
x, y =[], []
for i in np.arange(0,5):
    for j in np.arange(0,5):
        x.append(i)
        y.append(j)
ax[1].plot(x,y, "o")
ax[1].set_xlabel("URA")
ax[1].get_yaxis().set_visible(False)


x, y =[], []

for a in np.arange(0,2*np.pi,np.deg2rad(15)):
    x.append(np.sin(a)+1)
    y.append(np.cos(a))

ax[2].plot(x,y, "o")
ax[2].set_xlabel("UCA")
ax[2].get_yaxis().set_visible(False)

```

```python
#ds = data[data["capture"]  == "poli"]
ds= pd.read_json("../tw2/tw2.json")
x =[]
y=[]

for angle in set(ds["angle"]):
    for distance in set(ds[ds["angle"] ==angle]["distance"]):
        x.append(int(distance) * np.sin(np.deg2rad(angle)))
        y.append(int(distance) * np.cos(np.deg2rad(angle)))

plt.figure(figsize=(5,4))
plt.plot(x,y,".")
plt.plot(0,0,"ok")
plt.gca().invert_xaxis()

plt.ylabel("Distância (cm)")
plt.xlabel("Distância (cm)")
```
