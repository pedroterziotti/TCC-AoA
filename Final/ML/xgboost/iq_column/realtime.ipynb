```python
import paho.mqtt.client as mqtt
import json
import numpy as np
import xgboost as xgb
from numpy import linalg
import socket
from scipy.stats import zscore, mode
```

# Regressor

```python
model = xgb.XGBRegressor()
model.load_model("./xbg_regressor.json")

# Callback chamado quando o cliente recebe a resposta CONNACK do servidor.
def on_connect(client, userdata, flags, rc):
    print("Conectado com o código de resultado "+str(rc))
    client.subscribe("/gw/ac233fc09283/status")
    
    
# Callback chamado quando uma mensagem PUBLISH é recebida do servidor.
#antenna array configuration: 2,2,0,5,6,1,12,14

def on_message(client, userdata, msg):
    
    data = json.loads(msg.payload)
        
    iqs = np.array([data[1]["aoa"]["iq"][i:i+6] for i in range(16,len(data[1]["aoa"]["iq"]),6)][:-1])
    print(iqs)
    model_predictions= []

    for elemnt in iqs:
        model_predictions.append(model.predict(elemnt.reshape(1,6)))
    model_predictions=np.array(model_predictions)
    #print(f"zscore={zscore(model_predictions)}")
    print(model_predictions)
#    print(f"Media {np.mean(model_predictions)} \t Media filtrada {np.mean(model_predictions[abs(zscore(model_predictions)) <1])}")


client = mqtt.Client()
client.on_connect = on_connect
client.on_message = on_message

# client.connect("broker.hivemq.com", 1883, 60)
client.connect("test.mosquitto.org", 1883, 60)

client.loop_forever()
```
