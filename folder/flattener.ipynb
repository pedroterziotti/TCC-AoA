```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def single_antenna_to_dataframe(read_fp: str, write_fp=None) -> pd.DataFrame:
    '''Transforma uma captura com uma única antena em um df '''
    raw =pd.read_json(read_fp)
    df= list()
    for group in raw[1]:
        df.append(group)
    df = pd.DataFrame(df)
    df = df.join(pd.DataFrame(df["aoa"].to_list()))
    df.pop("aoa")
    df.pop("mac")
    df.pop("no")
    df.pop("raw")

    if write_fp is not None:
        df.to_json(write_fp)
    return df
```

```python
data =pd.DataFrame()
for ANTENNA_ORDER in os.listdir("./data"):
    for ANGLE in os.listdir(f"./data/{ANTENNA_ORDER}"):
        for FILE in os.listdir(f"./data/{ANTENNA_ORDER}/{ANGLE}"):
            try:
                df = single_antenna_to_dataframe(f"./data/{ANTENNA_ORDER}/{ANGLE}/{FILE}")
                df["Switch Order"] = ANTENNA_ORDER
                df["Angle"]= ANGLE
                df["Antenna"] = FILE.split("-")[0][-1]
                df["TAG"] = FILE.split("-")[1][-1]
                df["Entrada"] = FILE.split("-")[2].split(".")[0][-1]

                data = pd.concat([data,df])
            except Exception as e:
                print(f"./data/{ANTENNA_ORDER}/{ANGLE}/{FILE}")
                print(e.args)
data
```

```python
data = data.reset_index()
data.to_json("./data.json")
data
```
